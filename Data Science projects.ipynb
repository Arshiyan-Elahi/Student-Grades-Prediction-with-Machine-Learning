```python
import pandas as pd
import plotly.express as px

# Load dataset
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/student-mat.csv")

# Scatter Plot: Relationship between G1 and G3
scatter_fig = px.scatter(data, x="G1", y="G3", title="Scatter Plot: G1 vs G3 (First Exam vs Final Exam)")

# Bar Chart: Average G3 Score per Study Time
bar_fig = px.bar(data.groupby("studytime")["G3"].mean().reset_index(),
                 x="studytime", y="G3", title="Average Final Grade (G3) vs Study Time")

# Line Chart: G1, G2, and G3 Trends
line_fig = px.line(data, x=data.index, y=["G1", "G2", "G3"], title="Trend of G1, G2, and G3 Over Time")

# Histogram: Distribution of Final Grades (G3)
hist_fig = px.histogram(data, x="G3", nbins=10, title="Distribution of Final Grades (G3)")

# Box Plot: Study Time vs Final Grade (G3)
box_fig = px.box(data, x="studytime", y="G3", title="Study Time vs Final Grades (G3)")

# Display all graphs
scatter_fig.show()
bar_fig.show()
line_fig.show()
hist_fig.show()
box_fig.show()
```
